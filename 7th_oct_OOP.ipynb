##### Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.

```python
# create a class for bank account
class Bank_account:
    # create a class attribute to store all bank account objects
    all_accounts = []
    # initialize a constructor which takes 3 formal parameters - account name, accountr number, balance
    def __init__(account, account_name:str, account_number:int, balance:float):
        # create a instance variable to store account number
        account.account_number = account_number
        # create a instance variable to store account name
        account.account_name = account_name
        # create a instance variable to store account balance
        account.balance = balance
        # append the current object in class attribute list 
        Bank_account.all_accounts.append(account)
    # create a method to diposit amount in bank account
    def deposit_money(account, amount):
        # create a try block coz user can put negative value to deposit
        try:
            # if amount is greater than eq. to 0 then add it to balance
            if amount>0: 
                account.balance += amount
            # if amount is negative , raise a value error with appropriate message.
            else:
                raise ValueError("Enter a valid positive amount")
        # catch the value error in execpt block and print the message passed with execption
        except ValueError as v:
            print(v)
        # create another except block to catch errors which are anamolous to admin and print the message
        except Exception as E:
            print(E)
    # create a method to withdraw money
    def withdraw_money(account, amount):
        # create a try block coz there may be a possibility that balance < amount to be withdrawn
        try:
            # if balance is greater than or eq. to amount to be withdrawn
            if account.balance>=amount :
                # withdraw the money and decrease the balance.
                account.balance -= amount
            # otherwise balance < to amount to be withdrawn
            else:
                # raise the value error exception with an appropriate message.
                raise ValueError("Balance is not sufficient to withdraw")
        # catch the value error in execpt block and print the message passed with execption
        except ValueError as v:
            print(v)
        # create another except block to catch errors which are anamolous to admin and print the message
        except Exception as E:
            print(E)
    # initialize a repr Dunder method
    def __repr__(account) -> str:
        # return an appropriate message regarding current object
        return f'Bank Account [Account Owner Name : {account.account_name} , Account Number : {account.account_number} , Account Balance : {account.balance}]'
# create object for  Bank Account and pass user's name, account number, balance
account1 = Bank_account("Mr. Kelvin Clan", 1977700999936499, 0)
# call respective method to deposit method
account1.deposit_money(30000)
# call respective method to withdraw method
account1.withdraw_money(1000)
# create object for  Bank Account and pass user's name, account number, balance
account2 = Bank_account("Ms. Cristien Alfer", 19777009999712442, 0)
# call respective method to deposit method
account2.deposit_money(50000)
# call respective method to withdraw method
account2.withdraw_money(3000)
# create object for  Bank Account and pass user's name, account number, balance
account3 = Bank_account("Ms. Jennie Dikkens", 19777009999537821, 0)
# call respective method to deposit method
account3.deposit_money(90000)
# call respective method to withdraw method
account3.withdraw_money(800)
# print data regarding each account object in Bank Account class by usnig call attribute all_accounts
for i in Bank_account.all_accounts:
    print(i)



```

##### Problem 2: Employee Management Create a class representing an employee with attributes like employee ID, name, and salary. Implement methods to calculate the yearly bonus and display employee details.


```python
# create a class Employee
class Employee:
    # create a class variable to store data for all employees
    all_employees = []
    # initialize a constructor 
    def __init__(self, employee_id:str, name:str, salary:int):
        # initialize instance variable and store employee_id in it
        self.employee_id = employee_id
        # initialize instance variable and store name in it
        self.name = name
        # initialize instance variable and store salary in it
        self.salary = salary
        # append the current employee object in all_employee list
        Employee.all_employees.append(self)
    # create a function to display details of user
    def display_details(self):
        # return the data of that current object 
        return f'Details of Employee:\n Employee Id : {self.employee_id}\n Employee name : {self.name}\n Employee Salary : {self.salary}'
    # create a function to calculate  yearly bonus
    def get_yearly_bonus(self):
        # bonus is 8.33% of monthly salary
        bonus = self.salary *  8.33/100
        # multiply it by 12 to calculate yearly bonus
        yearly_bonus = bonus*12 
        # return the yearly bonus
        return f"Yearly bonus of {self.name} is"+str(yearly_bonus)
# create an object of employee and pass employee id, name and salary arguments.
employee1 = Employee('EPM0001',"Alferd Clan",40000)
# call display_details() method for this object
print(employee1.display_details())
# call get_yearly_bonus() method for this object
print(employee1.get_yearly_bonus())
# create an object of employee and pass employee id, name and salary arguments.
employee2 = Employee('EPM0002',"Jennifer Louis",30000)
# call display_details() method for this object
print(employee2.display_details())
# call get_yearly_bonus() method for this object
print(employee2.get_yearly_bonus())

```

##### Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.


```python
# create class for product 
class Products:
    # initialize a constructor
    def __init__(self) -> None:
        # create an instance dictionary to store products as keys and their quantity as values.
        self.all_products = dict()
    # define a function to add product and its quantity
    def add_product_quantity(self, product_name, product_quantity):
        # if product is already present then just update its quantity
        if product_name in self.all_products :
            self.all_products[product_name] += product_quantity 
        # if product is not present then add product and its quantity as key value pair in dictionary
        else:
            self.all_products[product_name] = product_quantity 
        # print the appropriate message
        print("Product quantity has been updated")
    # create a function to display the products
    def display_products(self):
        return self.all_products
# create a class to manage inventory and inherit the class product
class Inventory(Products):
    # define the constructor
    def __init__(self) -> None:
        super().__init__()
    # define a function to add the product
    def add_product(self, product_name):
        # add the product using add_product_quantity() method
        Inventory.add_product_quantity(self,product_name=product_name, product_quantity=1)
    # create  a method to update the product quantity
    def update_product_quantity(self,product_name, product_quantity):
        # update the quantity using add_product_quantity() method
        Inventory.add_product_quantity(self,product_name, product_quantity)
# create an object of Inventory()
inv1 = Inventory()
# display the products using display_products()
print("Products are : \n",inv1.display_products())
# add products and update quantity using add_product() method and update_product_quantity() method
inv1.add_product('product6')
inv1.update_product_quantity('product6',67)
inv1.update_product_quantity('product1',23)
# display the products using display_products()
print("Products are : \n",inv1.display_products())


# create an object of Inventory()
inv2 = Inventory()
# display the products using display_products()
print("Products are : \n",inv2.display_products())
# add products and update quantity using add_product() method and update_product_quantity() method
inv2.add_product('product6')
inv2.update_product_quantity('product6',26)
inv2.update_product_quantity('product1',32)
# display the products using display_products()
print("Products are : \n",inv2.display_products())
```

##### Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.



```python
# create the class for shape
class Shape:
    # initialize a constructor with arguments - length, width, height
    def __init__(self, length, width, height):
        # initialize the intance variables for length, width, height
        self.height = height
        self.width = width
        self.length = length
    # create the method to calculate area
    def get_area(self):
        # return the area of shape
        return f"The Area is  {self.length * self.width * self.height }"
    #  create the method to calculate perimeter
    def get_perimeter(self):
        # return the perimeter of shape
        return f"The Perimeter is  {2 * (self.length + self.width)}"
    # initialize a magic method to get a string while printing object
    def __repr__(self) -> str:
        # return the appropriate string message.
        return f"shape : {self.length} X {self.width} X {self.height}"
# create an object with arguments for length, width, height
shape1 = Shape(2,3,5)
# print the object
print(shape1)
# print the perimeter
print(shape1.get_perimeter())
# print the area
print(shape1.get_area())

# create an object with arguments for length, width, height
shape2 = Shape(8,6,2)
# print the object
print(shape2)
# print the perimeter
print(shape2.get_perimeter())
# print the area
print(shape2.get_area())
```
