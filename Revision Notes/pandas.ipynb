### 2 types of indexing 
* Implicit Indexing - It is the internal indexing that pandas give the dataframe. It is also called integer index
* Explicit Indexing - It is the indexing that the user defines for the dataframe

### Merge and join operation

```python
df1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]
})
df2 =  pd.DataFrame({'key1':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]
}
)
df1, df2
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

```python
pd.merge(df1, df2, how = 'outer')
```

```python
pd.merge(df1, df2, how = 'inner')
```

```python
pd.merge(df1, df2, how = 'cross')
```

```python
pd.merge(df1, df2, how = 'left', left_on='key1', right_on = 'key4')
```

```python
df1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]},
                    index = ['a', 'b', 'c', 'd', 'e']
)
df2 =  pd.DataFrame({'key6':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]
},
index = ['a', 'b', 'h', 'i', 'j']
)
```

```python
pd.merge(df1, df2, how='left', left_on='key1', right_on='key6')
```

```python
df1.join(df2, how = 'left', ) #join happens on the indexes , column name should be not be common
```

```python
df1.join(df2, how = 'left' )
```

```python
df1.join(df2, how = 'outer')
```
