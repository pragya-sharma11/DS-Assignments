```python
plt.figure(figsize=(10,5)) #take the canvas for graph/plot
plt.subplot(1,2,1) # 1,2,1 ---> on 1 page, make 2 subparts of that page and access 1st sub-page
plt.plot(x, np.sin(x),'.-')
plt.subplot(1,2,2) # 1,2,1 ---> on 1 page, make 2 subparts of that page and access 2nd sub-page
plt.plot(x, np.cos(x),'.')
```

```python
plt.figure(figsize=(10,5)) #take the canvas for graph/plot
# matrix = 2X1
plt.subplot(2,1,1) # 1,2,1 ---> on 1 page, make 2 subparts i.e. 2 rows and access 1st sub-page
plt.plot(x, np.sin(x),'.-')
plt.subplot(2,1,2) # 1,2,1 ---> on 1 page, make 2 subparts i.e. 2 rows  and access 2nd sub-page
plt.plot(x, np.cos(x),'.')
```

```python
plt.figure() #take the canvas for graph/plot
# here, matrix canvas will be of 2X2
plt.subplot(2,2,1) # 2,2,1 ---> on canvas make 2 rows, make 2 subparts of divided page and access 1st sub-page
plt.plot(x, np.sin(x),'.-')
plt.subplot(2,2,2) # 2,2,1 ---> on canvas make 2 rows, make 2 subparts of divided page and access 2nd sub-page
plt.plot(x, np.cos(x),'.')
plt.subplot(2,2,3) # 2,2,1 ---> on canvas make 2 rows, make 2 subparts of divided page and access 3rd sub-page
plt.plot(x, np.tan(x),'--')
plt.subplot(2,2,4) # 2,2,1 ---> on canvas make 2 rows, make 2 subparts of divided page and access 4rth sub-page
plt.plot(x, np.tanh(x))
```

```python
plt.figure(figsize=(10,11)) #take the canvas for graph/plot
# here, matrix canvas will be of 3X2
plt.subplot(3,2,1) # 2,2,1 ---> on canvas make 2 rows, make 2 subparts of divided page and access 1st sub-page
plt.plot(x, np.sin(x),'.-')
plt.subplot(3,2,2) # 2,2,1 ---> on canvas make 2 rows, make 2 subparts of divided page and access 2nd sub-page
plt.plot(x, np.cos(x),'.')
plt.subplot(3,2,3) # 2,2,1 ---> on canvas make 2 rows, make 2 subparts of divided page and access 3rd sub-page
plt.plot(x, np.tan(x),'--')
plt.subplot(3,2,4) # 2,2,1 ---> on canvas make 2 rows, make 2 subparts of divided page and access 4rth sub-page
plt.plot(x, np.tanh(x))
plt.subplot(3,2,5) # 2,2,1 ---> on canvas make 2 rows, make 2 subparts of divided page and access 3rd sub-page
plt.plot(x, np.tan(x),'--')
plt.subplot(3,2,6) # 2,2,1 ---> on canvas make 2 rows, make 2 subparts of divided page and access 4rth sub-page
plt.plot(x, np.tanh(x))
```

**`rgb format` ***

```python
plt.plot(x, np.sin(x),'--', color = (1.0,0.2,0.6))  # rgb format
```

```python
plt.plot(x, np.sin(x-0),'--', color = (0.4,0.2,0.4))
plt.xlim(1,15)
plt.ylim(1.5,-1.5)
plt.show()
```

```python
plt.plot(x, np.sin(x-0),'--', color = (0.4,0.2,0.4))
plt.xlim(1,15)
plt.ylim(1.5,-1.5)
plt.axis('equal') # to equalize the empty space from all 4 sides of graph  'to make a attractive plot for business team'
plt.show()
```
